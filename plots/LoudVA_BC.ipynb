# The Best Configurations for the LoudVA

In this notebook, we will find the best configurations for the LoudVA system. We will use the pareto front for each device to find the best configurations. We will use the plot function we created in ihelper.py to plot the pareto front for each device. We will then find the best configurations for the whole system by creating a pareto front for the whole system.

Remember, the pareto front is a set of configurations that are not dominated by any other configuration. In other words, the pareto front is a set of configurations that are the best in some way. The pareto front maximizes the throughput, minimizes the latency, and minimizes the energy consumption.

```python
import ihelper as i

devices = ['agx-xavier-00', 'xavier-nx-00', 'LoudJetson0']
measurement_code = 'Representative'
measurement_dirs = [f'/home/louduser/LoudVA/measurements/archive/{measurement_code}/{device}/measurements' for device in devices]
```

## The Pareto Front

### Throughput vs Energy with Latency Pareto Optimization (LoudVA)

```python
device_labels = ['AGX', 'NX', 'Nano']
i.plot(
    folder_paths=[f"{measurement_dir}/energy/freqs" for measurement_dir in measurement_dirs],
    directory_labels=device_labels,
    row_number_x=3,
    row_number_y=4,
    row_name_x='Throughput (Inference/sec)',
    row_name_y='Energy (J)',
    title='Throughput vs Energy with Latency Pareto Optimization (LoudVA)',
    connect_points=False,
    label_column=0,
    label_points=True,
    row_number_z=2,
    pareto_boundary=True,
    maximize_x=True,
    maximize_y=False,
    maximize_z=False,
    distrust_file_paths=[f"{measurement_dir}/retried_freqs.txt" for measurement_dir in measurement_dirs],
    distrust_threshold=0,
    debug_mode=False,
    export_file_path=f'/home/louduser/LoudVA/measurements/archive/{measurement_code}/pareto.csv'
)
```
